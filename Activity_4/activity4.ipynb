# Import libraries and declare variables

```python
import control as ctrl
import matplotlib.pyplot as plt

num1 = [1]
den1 = [1, 2, 2]

num2 = [1]
den2 = [1, 2, 1]

G1 = ctrl.tf(num1, den1)
G2 = ctrl.tf(num2, den2)

poles1 = ctrl.poles(G1)
poles2 = ctrl.poles(G2)

zeros1 = ctrl.zeros(G1)
zeros2 = ctrl.zeros(G2)
```

# Problem #1

```python
print(G1)
print("") #spacing
print("--------------------")
print("") #another spacing
print("Poles:", poles1)
print("--------------------")
print("") #another spacing
print("Zeros:", zeros1)

ctrl.pzmap(G1, grid=True)
plt.show()
```

# Problem #2

```python
print(G2)
print("") #spacing
print("--------------------")
print("") #another spacing
print("Poles:", poles2)
print("--------------------")
print("") #another spacing
print("Zeros:", zeros2)

ctrl.pzmap(G2, grid=True)
plt.show()
```
