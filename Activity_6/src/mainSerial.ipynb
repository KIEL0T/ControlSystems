```python
import serial
import time
```

```python
arduino = serial.Serial(port='COM3', baudrate=9600, timeout=1)
time.sleep(2)
```

```python
while True:
    command = input("Enter 1 to turn on LED, 0 to turn off LED, q to quit: ")

    if command == '1':
        arduino.write(b'1')
    elif command == '0':
        arduino.write(b'0')
    elif command ==  'q':
        break
    else:
        print("Invalid Command")
        
arduino.close()
```
